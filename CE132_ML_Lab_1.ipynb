```python
import numpy as np
import random as rand

x = [1, 2, 3]
print(x, "Type: ", type(x))

arr = np.array(x)
print(arr, "Type: ", type(arr))

print("Sum of list: ", x + x) # It will concatenate the list
print("Sum of array: ", arr + arr) # Addition of two list will be performed

x = ["Harsh", "Sojitra"]
print(x * 2) # ['Harsh', 'Sojitra', 'Harsh', 'Sojitra']

arr1 = np.array([[1, 2, 3], [3, 4, 5]])
arr2 = np.array([[1, 5], [2, 6], [1, 7]])

# size must be same of both matrix for addition
# print(arr1 + arr2)
# print(np.add(arr1, arr2))

# no. of rows in one matrix and no. of columns in another matrix must be same
print(np.dot(arr1, arr2))
print(arr1.dot(arr2))
print(arr1 @ arr2)

print(arr1.T)

arr1 = np.array([[1, 9], [2, 5]])
arr2 = np.array([[2, 6], [1, 4]])

# 1 9     2 6
#      *
# 2 5     1 4

print(arr1 @ arr2)
print(arr1 * arr2)

print(np.linalg.norm(arr1))

arr2 = np.array([["Harsh Sojitra"]])
for a, b  in zip(arr1, arr2):
  print(a, b)
```
